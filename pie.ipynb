
# pie(x)
Plot a pie chart.

See `~matplotlib.axes.Axes.pie`.


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))
explode=(0, 0, 0, 0.15)

# plot

fig, ax = plt.subplots()
ax.pie(x,  
       colors=colors, 
       explode=explode, 
       radius=3, 
       center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, 
       autopct='%1.1f%%', 
       startangle=90, 
       frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

# plt.pie(x, 
#        colors=colors, 
#        center=(4,4), 
#        explode=explode, 
#        wedgeprops={"linewidth": 1, "edgecolor": "white"}, 
#        autopct='%1.1f%%', 
#        startangle=90, 
#        frame=True)

#plt.axis('equal')

plt.show()
```
